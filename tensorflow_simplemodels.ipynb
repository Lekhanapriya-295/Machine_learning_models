```python
import tensorflow as tf
import numpy as np
from tensorflow import keras
print(tf.__version__)
model=keras.Sequential([keras.layers.Dense(units=1,input_shape=[1])])
model.compile(optimizer='sgd',loss='mean_squared_error',metrics=['accuracy'])
xs=np.array([-1.0,0.0,1.0,2.0,3.0,4.0],dtype=float)
ys=np.array([-3.0,-1.0,1.0,3.0,5.0,7.0],dtype=float)
model.fit(xs,ys,epochs=1500)
print(model.predict(np.array([10.0])))
```

```python
#House rent according to BHK
import tensorflow as tf
import numpy as np
from tensorflow import keras
print(tf.__version__)
model=keras.Sequential([keras.layers.Dense(units=1,input_shape=[1])])
model.compile(optimizer='sgd',loss='mean_squared_error',metrics=['accuracy'])
xs=np.array([1,2,3,4,5],dtype=float)#BHK
ys=np.array([8,15,12,30,40],dtype=float)#rent in thousands
model.fit(xs,ys,epochs=500)
pred=model.predict(np.array([3.5]))
print(pred)
```

```python
#salary based on experience
import tensorflow as tf
import numpy as np
from tensorflow import keras
model=keras.Sequential([keras.layers.Dense(units=1,input_shape=[1])])
model.compile(optimizer='sgd',loss='mean_squared_error')
xs=np.array([1,2,3,4,5],dtype=float)#years of experience
ys=np.array([3,5,7,9,11],dtype=float)#salaries in lakhs
model.fit(xs,ys,epochs=1000)
print(model.predict(np.array([6])))
```

```python
#electricity bill base on units
import tensorflow as tf
import numpy as np
from tensorflow import keras
model=keras.Sequential([keras.layers.Dense(1,input_shape=[1])])
model.compile(optimizer='adam',loss='mean_squared_error')
xs=np.array([100,200,300,400,500],float)
ys=np.array([500,1000,1500,2000,2500],float)
model.fit(xs,ys,epochs=300)
print("bill for 350 units:")
print(model.predict(np.array([350.0])))
```
